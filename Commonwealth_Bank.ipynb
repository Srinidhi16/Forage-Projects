```python
#Removing names
df = df.drop(columns = ['name'])
```

```python
#Removing address
df = df.drop(columns = ['address'])
```

```python
#Removing bithdate and unnamed
df = df.drop(columns = ['birthdate','Unnamed: 0'])
```

```python
# Creating a function to mask credit card no.
import re

def mask_credit(num):    ###ph can be string / Nan
  if pd.isna(num):        #If value is missing
    return num
  digits = re.sub(r'\D', "",str(num))    #str : converts the value to a string \D ;→ regex pattern meaning “anything that is NOT a digit”
  masked = '*' * (len(digits)-4) + digits[-4:]
  return masked
```

```python
#Apply mask_credit function to credit card column
df['credit_card_number'] = df['credit_card_number'].apply(mask_credit)
```

```python
#Removing credit card security code and credit card provider as that isn't required
df = df.drop(columns = ['credit_card_security_code'])
```

```python
#removing current location lat & long
df = df.drop(columns = ['current_location'])
```

```python
#From residence columns extract state and pin-code
df['pincode'] = df['residence'].str[-5:]  #.str applies string slicing element-wise to the column
```

```python
df['state'] = df['residence'].str[-8:-6]
```

```python
df.head()
```

```python
#Drop column residence
df = df.drop(columns = ['residence'])
```

```python
#Func to mask email

def mask_mail(email):
  if pd.isna(email):
    return email
  mail = str(email)

  #split user and domain
  user, domain = mail.split('@')
  #split domain name and extension
  domain_name, domain_ext = domain.split('.', 1)

  masked_user = user[0] + '*' * (len(user)-1)
  masked_domain = domain[0] + '*' * (len(domain)-1)

  return f"{masked_user}@{masked_domain}.{domain_ext}"
```

```python
df['email'] = df['email'].apply(mask_mail)
```

```python
df.head()
```

```python
def age_range(age,step=5):
  if pd.isna(age):
    return age
  age = int(age)
  lower = (age // step) * step + 1
  upper = lower + step - 1

  return f"{lower}-{upper}"
```

```python
df['age_range'] = df['age'].apply(age_range)
```

```python
#Remove age column after deriving age_range
df = df.drop(columns = ['age'])
```
