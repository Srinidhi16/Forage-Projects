```python
import pandas as pd
df = pd.read_excel('/content/mobile_customers.xlsx')
```

**Required columns for analysis:**

customer_id, date_registered, gender, email, employer, age bucket and salary range, credit card masked no., expiry

```python
#Removing names
df = df.drop(columns = ['name'])
```

```python
#Removing address
df = df.drop(columns = ['address'])
```

```python
#Removing bithdate and unnamed
df = df.drop(columns = ['birthdate','Unnamed: 0'])
```

```python
# Creating a function to mask credit card no.
import re

def mask_credit(num):    ###ph can be string / Nan
  if pd.isna(num):        #If value is missing
    return num
  digits = re.sub(r'\D', "",str(num))    #str : converts the value to a string \D ;→ regex pattern meaning “anything that is NOT a digit”
  masked = '*' * (len(digits)-4) + digits[-4:]
  return masked
```

```python
#Apply mask_credit function to credit card column
df['credit_card_number'] = df['credit_card_number'].apply(mask_credit)
```
